```python
def scrape_flipkart_urls(product_category_url,max_product_per_category):
    driver.get(product_category_url)
    time.sleep(random.uniform(3, 5))  # Let the page load

    urls = []
    page_number = 1
    unique_products = set()
    
    while len(unique_products) < max_product_per_category and page_number < 12:
        time.sleep(random.uniform(3, 5))
        paginated_url = f"{product_category_url}&page={page_number}"
        driver.get(paginated_url)
    
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        # Find all review containers
        products = soup.find_all('a', {'class': 'CGtC98'}) or soup.find_all('a', {'class': 'WKTcLC'})
        page_number +=1
        for product in products:
            try:
                href = 'https://www.flipkart.com'+product.get('href')
                name = product.find('div', {'class': 'KzDlHZ'}).text.split('(')[0].strip() or product.get('title')
                    

                if name not in unique_products:
                    urls.append(href)
                    unique_products.add(name)
                    
            except Exception as e:
                print("Error extracting urls, skipping:", e)

    return urls
```

```python

def scrape_multiple_flipkart_products_urls(product_category_urls, max_product_per_category=20):

    for index, product_category_url in enumerate(product_category_urls):
        print(f"Scraping product {index + 1}/{len(product_category_urls)}")
        
        product_url = scrape_flipkart_urls(product_category_url, max_product_per_category)
        product_urls.extend(product_url)

        
        df = pd.DataFrame(product_url)
        
        df.to_csv('flipkart_products_urls.csv', mode='a', header=False, index=False)

    print(f"Scraping complete. Total urls scraped: {len(product_urls)}")
    driver.quit()
```

```python
def scrape_flipkart_reviews(product_url, max_reviews , order   , max_page ):
    driver.get(product_url)
    time.sleep(random.uniform(2, 4))  

    
    try:
        review_button = driver.find_element(By.CLASS_NAME, "_23J90q.RcXBOT") 
        review_button.click() 
        time.sleep(random.uniform(2, 4)) 
        select_element = driver.find_element(By.CLASS_NAME, "OZuttk.JEZ5ey")

        select = Select(select_element)

        select.select_by_value(order)
        time.sleep(random.uniform(2, 4)) 
    except Exception as e:
        print("Failed to click the reviews button:")
        return []
    all_reviews_url = driver.current_url
    reviews_list = []
    review_count = 0
    page_number = 1

    while review_count < max_reviews and page_number < max_page:
        time.sleep(random.uniform(2, 4))
        paginated_url = f"{all_reviews_url}&page={page_number}"
        driver.get(paginated_url)

        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        reviews = soup.find_all('div', {'class': 'col EPCmJX Ma1fCG'})
        page_number +=1
        for review in reviews:
            try:
                rating = review.find('div', {'class': 'XQDdHH Ga3i8K'}) or \
                  review.find('div', {'class': 'XQDdHH Js30Fc Ga3i8K'}) or \
                        review.find('div', {'class': 'XQDdHH Czs3gR Ga3i8K'}) 

                rating = rating.text.strip()
                review_text = review.find('div', {'class': 'ZmyHeo'}).find('div').find('div').text.strip() 

                reviews_list.append({
                    'rating': rating,
                    'review_text': review_text
                })

                review_count += 1
                if review_count >= max_reviews:
                    return reviews_list
            except Exception as e:
                print("Error extracting review, skipping:", e)

    return reviews_list
```

```python

def scrape_multiple_flipkart_products(product_urls, max_reviews_per_product , order = "MOST_RECENT"  , max_page = 13):
    all_reviews = []

    for index, product_url in enumerate(product_urls):
        print(f"Scraping product {index + 1}/{len(product_urls)}")
        
        product_reviews = scrape_flipkart_reviews(product_url, max_reviews=max_reviews_per_product , order=order  , max_page=max_page )
        all_reviews.extend(product_reviews)

        df = pd.DataFrame(product_reviews)
        df.to_csv('flipkart_reviews.csv' , mode='a' , header=False, index=False)

    print(f"Scraping complete. Total reviews scraped: {len(all_reviews)}")
    driver.quit()

```
